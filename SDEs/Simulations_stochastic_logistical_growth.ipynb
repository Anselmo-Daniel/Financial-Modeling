```python
import numpy as np
from matplotlib import pyplot as plt
```

# Questão 1

Vamos usar o método de euler para resolver a seguinte EDO:

$$ \frac{dx}{dt} = (\alpha -\beta x)x $$

$$ x(0) = x_{0}$$

```python
### Parametros

alpha = 1
beta  = 1
x0 = 1e-2  # 0 < x0 < alpha/beta, t0 = 0 e 0 << delta_t << T
```

```python
n = 100
a, b = 0, 10
delta_t = (b - a)/n
eixo_x = np.linspace(a, b, num = n, endpoint = True)

eixo_y1 = np.zeros(len(eixo_x))
eixo_y1[0] = x0
```

```python
for i in range(n-1):
  eixo_y1[i+1] = eixo_y1[i] + ( eixo_y1[i] * (1- eixo_y1[i])) * delta_t
```

```python
plt.plot(eixo_x, eixo_y1)
plt.show()
```

# Questão 2

Vamos usar o método de euler para resolver :

$$ \frac{dx}{dt} = (\alpha -\beta X_t)X_t $$

<br>

$$ X(0) = N(x_0, \sigma_{x_0}^{2}) $$

e

Alpha e Beta tem distribuições Normais de parametros : $ N(x_\alpha, \sigma_\alpha)$ e $N(x_\beta, \sigma_\beta) $

```python
alpha = np.random.normal(0, 1, n)*1e-2 + 1
beta  = np.random.normal(0, 1, n)*1e-2 + 1
x0    = np.random.normal(0, 1, n)*1e-3 + 1e-2

for i in range(100):

  eixo_y2 = np.zeros(len(eixo_x))
  eixo_y2[0] = x0[i]

  for j in range(n-1):
    eixo_y2[j+1] = eixo_y2[j] + ( eixo_y2[j] * (alpha[i] - beta[i]*eixo_y2[j])) * delta_t

  plt.plot(eixo_x, eixo_y2, color = 'red')
```

# Questão 3

Vamos usar o método de euler para resolver :

$$ \frac{dx}{dt} = (A_t -b X_t)X_t $$

<br>

$$ X(0) = x0 $$

Onde $ A_t = a + \sigma sin(W_t) $ e $ W_{t+\tau} - W_t $ tem distribuição $ N(0, \tau) $

```python
wt = 0
beta  = 1
x0    = 1e-2

for i in range(100):
  eixo_y3 = np.zeros(len(eixo_x))
  eixo_y3[0] = x0

  for j in range(n-1):
    wt += np.random.normal(0, 1)* np.sqrt(delta_t)
    alpha = 1 + 0.2*np.sin(wt)
    eixo_y3[j+1] = eixo_y3[j] + ( eixo_y3[j] * (alpha - beta*eixo_y3[j])) * delta_t

  plt.plot(eixo_x, eixo_y3, color = 'red', alpha = 0.2)
```

# Questão 4

Vamos usar o método de euler para resolver :

$$ {dx} = (\alpha -\beta X_t)X_t dt + \sigma X_t dW_t $$

<br>

$$ X(0) = x0 $$

$ W_{t+\tau} - W_t $ tem distribuição $ N(0, \tau) $

```python
alpha = 1
beta  = 1
x0    = 1e-2
sigma = 0.5

for i in range(100):
  eixo_y4 = np.zeros(len(eixo_x))
  eixo_y4[0] = x0

  for j in range(n-1):
    dw = np.random.normal(0, np.sqrt(delta_t))

    eixo_y4[j+1] = eixo_y4[j] + ( eixo_y4[j] * (alpha - beta*eixo_y4[j]) + sigma*eixo_y4[j]*dw ) * delta_t

  plt.plot(eixo_x, eixo_y4, color = 'red', alpha = 0.2)
```
